```python
import pandas;
```

```python
countries = pandas.read_csv('countries.csv');
countries
```

```python
zambia = countries[countries.country == 'Zambia'];
zambia
```

```python
from matplotlib import pyplot;

pyplot.plot(zambia.year, zambia.gdpPerCapita);
pyplot.title('Zambia\'s GDP per capita');
pyplot.show();
```
